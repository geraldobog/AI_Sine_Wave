```python
#imports
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
t = np.arange(200).reshape(-1,1)/50 #input t arranged into an array from 0 to 200 and reshaped from -1 to 1
y = np.sin(3*t)#output function sin(3t)
```

```python
model = Sequential()#initialize the sequential model from keras
model.add(Dense(40, input_dim=1, activation='relu'))#first input layer with 40
model.add(Dense(20, activation='relu'))#first 
model.add(Dense(10, activation='relu'))#second
model.add(Dense(1))#output mapped to a single predicted y value
model.compile(loss='mean_squared_error', optimizer='adam')#metrics used mean sqr error and used adam optimizer
model.fit(t, y, epochs=1000, batch_size=50)#modelling with 1000 epochs
```

```python
predictions = model.predict(t)
print(predictions)
# print the model's predictions
```

```python
#plotting sin(3t)
import matplotlib.pyplot as plt

plt.plot(t, y, 'b')
plt.title('y = $sin(3t)$')
plt.ylabel('Real y')


```

```python
#plotting predicted sin(3t) for 1000 epochs
import matplotlib.pyplot as plt

plt.plot(t,predictions, 'b')
plt.xlabel('t')
plt.ylabel('Predicted y')
plt.title('predicted sin(3t) after 1000 epochs')

plt.show()
```
